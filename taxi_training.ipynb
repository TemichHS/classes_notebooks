```python
import pandas as pd
```

```python
taxi = pd.read_csv('2_taxi_nyc.csv')
```

```python
taxi.shape
```

```python
taxi.info()
```

```python
taxi = taxi.rename(columns={'pcp 01': 'pcp_01', 'pcp 06': 'pcp_06', 'pcp 24': 'pcp_24'})
```

```python
taxi.head()
```

```python
taxi \
    .groupby('borough') \
    .pickups \
    .sum() \
    .sort_values(ascending=False)
```

```python
taxi  \
    .pickups \
    .idxmin()
```

```python
taxi \
    .groupby('borough') \
    .pickups \
    .sum() \
    .idxmin()
```

```python
taxi \
    .groupby(['borough', 'hday'], as_index=False) \
    .pickups \
    .mean() \
    .round(2)
```

```python
taxi.head(3)
```

```python
taxi \
    .groupby(['borough', 'pickup_month'], as_index=False) \
    .pickups \
    .sum() \
    .sort_values(by='pickups', ascending=False)
```

```python
def temp_to_celcius(x):
    return (x - 32) * 5.0 / 9.0
```

```python
temp_to_celcius(taxi['temp'])
```
